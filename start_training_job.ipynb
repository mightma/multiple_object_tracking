## Launch Training Jobs,
Reference the local path on the SageMaker instance corresponding to the input S3 path as follows
![](https://docs.aws.amazon.com/images/sagemaker/latest/dg/images/sagemaker-training-storage.png)

```python
import time
from datetime import datetime

from sagemaker.pytorch import PyTorch

instance_count = 1
instance_type = 'ml.g5.12xlarge'

# Get the current time
current_time = datetime.now()

# Format the current time as a string
formatted_time = current_time.strftime("%Y%m%d_%H%M%S")
print(formatted_time)

training_dataset_s3_dir = "s3://clean-pro-multi-obejcts-tracking/private-dataset/1000_samples/train2017"
validation_dataset_s3_dir = "s3://clean-pro-multi-obejcts-tracking/private-dataset/1000_samples/val2017"
output_s3_dir = f's3://{sagemaker_default_bucket}/projects/multi-objects-tracking/train-{formatted_time}/'
print(f"The artifact will be saved at {output_s3_dir}")

environment = {
    'NODE_NUMBER': str(instance_count)
}


estimator = PyTorch(entry_point='entry.py',
                    source_dir='boxmot/',
                    base_job_name='clean-pro-multi-objects-tracking',
                    role=role,
                    environment=environment,
                    framework_version='2.1.0',
                    py_version='py310',
                    script_mode=True,
                    instance_count=instance_count,
                    instance_type=instance_type,
                    output_path=output_s3_dir)

estimator.fit(inputs={
    "train2017": training_dataset_s3_dir,
    "val2017": validation_dataset_s3_dir,
})
```
